### 1. The Identity Swap
**Goal:** Swap `x = 100` and `y = 200` in **exactly one line** – no temporary variable

```python
x, y = 100, 200
x, y = y, x
print(f"After swap → x = {x}, y = {y}")   # → x = 200, y = 100
```

### 2. The Type Auditor
**Goal:** Show that `input()` always returns a string and casting creates a new object

```python
raw = input("Enter a number: ")
print(f"Raw input type  : {type(raw)}")

try:
    number = float(raw)
    print(f"After float()   : {type(number)}")
    print(f"Value           : {number}")
except ValueError:
    print("Invalid input – cannot convert to float")
```

### 3. The Precision Banker
**Goal:** Print `100 / 3` formatted to exactly **3 decimal places** with a dollar sign

```python
bill = 100 / 3
print(f"${bill:.3f}")   # → $33.333
```

### 4. The Modulo Architect
**Goal:** Convert total seconds into Hours, Minutes, and Seconds using `//` and `%`

```python
total_seconds = int(input("Enter total seconds (e.g. 3665): "))

hours = total_seconds // 3600
remaining = total_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f"{hours} hours, {minutes} minutes, {seconds} seconds")
```
