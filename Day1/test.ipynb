## Day 1: The Environment & Building Blocks

### Micro-Challenge: The Identity Swap

```python
x, y = 100, 200
x, y = y, x
print(f"After swap: x = {x}, y = {y}")
```

### Micro-Challenge: The Type Auditor

```python
raw = input("Enter a number: ")
print("Raw input type:", type(raw))

try:
    number = float(raw)
    print("After casting to float:", type(number))
    print("Value:", number)
except ValueError:
    print("Error: Invalid number")
```

### Micro-Challenge: The Precision Banker

```python
bill = 100 / 3
print(f"${bill:.3f}")
```

### Micro-Challenge: The Modulo Architect

```python
total_seconds = int(input("Enter total seconds: "))

hours = total_seconds // 3600
remaining = total_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f"{hours} hours, {minutes} minutes, {seconds} seconds")
```

### Micro-Challenge: The Guard Clause (Short-Circuiting)

```python
user = None  # Try also with user = {"access": "admin"}

if user is not None and user.get("access") == "admin":
    print("Admin access granted")
else:
    print("Access denied or user is None")
```

## Day 3: Loops Iteration

### Micro-Challenge: The Infinite Guardian

```python
while True:
    password = input("Enter password (type 'stop' to exit): ")
    if password == "stop":
        print("Exiting...")
        break
    print("Password received:", password)
```

### Micro-Challenge: The Accumulator Pattern

```python
n = int(input("Enter N: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum from 1 to", n, "=", total)
```

### Micro-Challenge: The Efficient Skipper

## Day 4: Lists (Data Structures I)

### Micro-Challenge: The Reference Trap

```python
a = [1, 2, 3]
b = a
b[0] = 99
print("a:", a)  # Both change!
print("b:", b)
```

### Micro-Challenge: The Slicing Surgeon

```python
data = [10, 20, 30, 40, 50]
last_three_reverse = data[-1:-4:-1]  # or data[:1:-1]
print(last_three_reverse)  # [50, 40, 30]
```

### Micro-Challenge: The Stack Emulator

```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print("Popped:", stack.pop())
print("Stack now:", stack)
```

### Micro-Challenge: The One-Line Architect

```python
nums = list(range(1, 11))
even_squares = [x**2 for x in nums if x % 2 == 0]
print(even_squares)  # [4, 16, 36, 64, 100]
```

## Day 5: Dictionaries (Hash Maps)

### Micro-Challenge: The Safe Vault

```python
user = {"id": 1}
# print(user["email"])  # KeyError!
email = user.get("email", "No Email Found")
print(email)
```

### Micro-Challenge: The Frequency Counter

```python
text = "banana"
freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)  # {'b': 1, 'a': 3, 'n': 2}
```

### Micro-Challenge: The Database Merger

```python
defaults = {"theme": "light", "audio": "on"}
user_pref = {"theme": "dark"}

merged = defaults | user_pref  # Python 3.9+
# merged = defaults.copy(); merged.update(user_pref)  # Alternative

print(merged)  # {'theme': 'dark', 'audio': 'on'}
```

## Day 6: Functions Modularity

### Micro-Challenge: The Scope Fortress

```python
x = 10

def change_x():
    x = 20  # Local variable
    print("Inside function:", x)

change_x()
print("Outside function:", x)  # Still 10
```

### Micro-Challenge: The Pure Return

```python
def add(a, b):
    print(a + b)

res = add(5, 5)
print("Returned value:", res)  # None
```

### Micro-Challenge: The Default Gateway

```python
def connect(port=3306):
    print(f"Connecting to {port}")

connect()        # Uses default 3306
connect(5432)    # Custom port
```

### Micro-Challenge: The Logic Gate

```python
def is_even(num):
    return num % 2 == 0

print(is_even(4))  # True
print(is_even(7))  # False
```

## Day 7: Error Handling (Exceptions)

### Micro-Challenge: The Input Guard

```python
while True:
    try:
        age = int(input("Enter your age: "))
        print("Age accepted:", age)
        break
    except ValueError:
        print("Numbers only")
```

### Micro-Challenge: The Math Safety Net

```python
x = 0
try:
    print(100 / x)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

### Micro-Challenge: The Cleanup Crew

```python
try:
    result = 100 / 0
except ZeroDivisionError:
    print("Division failed")
finally:
    print("Execution Complete")
```

### Micro-Challenge: The Custom Signal

```python
try:
    num = float(input("Enter a positive number: "))
    if num < 0:
        raise ValueError("No negatives")
    print("Valid number:", num)
except ValueError as e:
    print(e)
```
